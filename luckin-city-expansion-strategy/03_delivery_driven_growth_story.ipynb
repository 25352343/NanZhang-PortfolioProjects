## Purpose of This Analysis

This analysis examines whether city and store growth
is primarily driven by delivery or offline consumption.

Understanding this is critical for determining
the appropriate expansion model.


## Business Question & Metrics

**Core Question**

What proportion of store GMV is generated via delivery?

**Key Metrics**

- Total monthly GMV per store
- Delivery GMV per store
- Delivery GMV share


## Analytical Approach

For each store and month in 22Q1:
- Aggregate total GMV
- Aggregate delivery GMV
- Calculate delivery share

SQL reference:
`sql/03_delivery_share_by_store.sql`


## Key Findings

- Some cities consistently show high delivery GMV share
- Delivery-driven stores tend to show more stable demand
- Offline-heavy stores exhibit higher volatility


## Strategic Implications

- Delivery-driven cities support faster and lighter expansion
- Offline-heavy cities require more precise site selection

Before scaling aggressively, one final validation is required:
**Can marketing investment reliably accelerate early-stage growth?**
