## Purpose of This Analysis

The first step of the expansion analysis is to identify
which cities in East China demonstrate **real growth potential**.

Rather than looking at absolute GMV levels,
this analysis focuses on **YoY growth after removing seasonality**,
which is more relevant for forward-looking expansion decisions.


## Business Question & Metrics

**Core Question**

Which cities achieved the fastest GMV growth in 22Q1
compared with 21Q1?

**Key Metrics**

- GMV (Gross Merchandise Value)
- YoY Growth Rate (22Q1 vs 21Q1)
- Number of active stores per city

**Why Q1 vs Q1?**

Comparing the same quarter removes seasonal effects
such as holidays and weather, making growth more comparable.


## Analytical Approach

City-level GMV is aggregated for:
- 21Q1 (Jan–Mar 2021)
- 22Q1 (Jan–Mar 2022)

The SQL logic used for this step can be found in:

`sql/01_city_growth_q1_yoy.sql`

All queries were executed directly in Alibaba Cloud (MaxCompute),
and results were reviewed online.


## Key Findings

Based on the YoY comparison:

- Several cities show strong GMV acceleration in 22Q1
- Not all high-GMV cities exhibit high growth
- Some smaller cities outperform larger ones in growth rate

This indicates that **expansion priority should not be determined
by size alone**, but by growth dynamics.


## Strategic Implications

Cities with strong YoY growth are shortlisted
for further structural analysis.

However, growth alone is not sufficient.
The next step is to examine **whether this growth is scalable**
at the store level.
