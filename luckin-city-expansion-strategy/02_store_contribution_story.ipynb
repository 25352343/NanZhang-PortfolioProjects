## Purpose of This Analysis

After identifying high-growth cities,
the next step is to evaluate **how that growth is generated**.

Specifically, we examine whether city-level growth
is driven by multiple stores or concentrated in a few flagship stores.


## Business Question & Metrics

**Core Question**

Is city growth driven by:
- diversified store clusters, or
- a small number of top-performing stores?

**Key Metrics**

- Store-level GMV (22Q1)
- GMV ranking within each city (Top 3 stores)


## Analytical Approach

For each city in 22Q1:
- Aggregate GMV at the store level
- Rank stores by GMV contribution

SQL reference:
`sql/02_store_gmv_ranking.sql`


## Key Findings

- Some high-growth cities show balanced GMV contribution
  across multiple stores
- Other cities rely heavily on one flagship store
- In extreme cases, the Top 1 store contributes
  a disproportionate share of city GMV


## Strategic Implications

- Cities with diversified store contribution
  are more suitable for aggressive expansion
- Flagship-driven cities require cautious replication
  and deeper understanding of success factors

For cities deemed scalable,
the next question is **what drives that scalability**.
