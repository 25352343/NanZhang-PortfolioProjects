## Purpose of This Analysis

The final step validates whether marketing investment
can effectively accelerate store performance,
especially during early expansion phases.


## Business Question & Metrics

**Core Question**

For stores with significant marketing investment,
does spend translate into real sales growth?

**Key Metrics**

- Weekly marketing spend
- Weekly SKU-level sales
- Top-selling products per store


## Analytical Approach

The analysis focuses on:
- Stores with weekly marketing spend > 500
- Corresponding weekly Top 5 SKUs by sales

SQL reference:
`sql/04_marketing_roi_store_sku.sql`


## Key Findings

- In effective cases, marketing spend amplifies
  already strong core SKUs
- In weaker cases, spend increases exposure
  without meaningful sales lift


## Strategic Implications

- Marketing investment should be used selectively
- SKU strength must be validated before scaling spend
- Expansion should avoid relying on marketing alone
